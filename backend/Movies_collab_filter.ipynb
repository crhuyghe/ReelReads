```python
import pandas as pd
import random
from sklearn.metrics.pairwise import cosine_similarity
user_item_matrix_path = "data/als_movie_ratings.csv"
user_item_matrix = pd.read_csv(user_item_matrix_path)
user_item_matrix
```

```python
import pandas as pd
import random

movie_path = "data/movies_metadata_trimmed.csv"
movie_df = pd.read_csv(movie_path)
movie_df
```

```python
ratings_df = pd.read_csv("data/ratings_trimmed.csv")
ratings_df = ratings_df.drop(columns=['timestamp'], errors='ignore')
ratings_df = ratings_df.movieId.unique()
ratings_df.tolist()
```

```python
NewMoviedf = movie_df[movie_df['id'].isin(ratings_df)]
NewMoviedf
```

```python
FantasyDf = NewMoviedf[NewMoviedf['genres'].apply(lambda x: 'Fantasy' in x)]
FantasyDf

# x = fantasy_movies.loc[33]

# # print out all of the genres
# print(x['genres'])
```

```python
selected_magic_books = FantasyDf.sample(n=15, random_state=42)  # Select 5 random books
selected_movie_id = selected_magic_books["id"].astype(str).tolist()
selected_movie_id
```

```python
for movie_id in selected_movie_id:
    print(NewMoviedf.loc[NewMoviedf['id'] == int(movie_id)]['original_title'])
# print(books_df.loc[books_df['Id'] == 702817]['Name'])
```

```python
new_user_vector = pd.DataFrame(0, index=[0], columns=user_item_matrix.columns, dtype=float)
for movie_id in selected_movie_id:
    new_user_vector.loc[0, movie_id] = round(random.uniform(4.0, 5.0), 2)
new_user_vector
```

```python
updated_user_item_matrix = pd.concat([new_user_vector, user_item_matrix], ignore_index=True)
updated_user_item_matrix
```

```python
new_user_vector_values = new_user_vector.values
cosine_sim = cosine_similarity(new_user_vector_values, updated_user_item_matrix.values)
cosine_sim
```

```python
# Convert similarity scores to DataFrame
similarity_df = pd.DataFrame({
    'userId': updated_user_item_matrix.index,
    'cosine_similarity': cosine_sim.flatten()
}).sort_values(by='cosine_similarity', ascending=False)

similarity_df
```

```python
similarity_df = similarity_df[similarity_df['userId'] != test_user_id]
similarity_df
```

```python
top_similar_users = similarity_df.head(10)['userId'].tolist()
top_similar_users
```

```python
# Find books rated by similar users but not by the test user
test_user_rated_movies = set(updated_user_item_matrix.loc[test_user_id][updated_user_item_matrix.loc[test_user_id] > 0].index)
similar_users_ratings = updated_user_item_matrix.loc[top_similar_users]
similar_users_ratings
```

```python
# Find books that similar users rated but the test user hasn't rated
recommendation_candidates = similar_users_ratings.drop(columns=test_user_rated_movies, errors='ignore')

# Calculate average ratings of these books from similar users
recommendations = recommendation_candidates.mean().reset_index()
recommendations.columns = ['movie_id', 'avg_rating']

# Sort recommendations by highest average rating
recommendations = recommendations.sort_values(by='avg_rating', ascending=False)
recommendations
```

```python
# Compute a weighted rating based on similarity scores
weighted_ratings = (similar_users_ratings.T * similarity_df['cosine_similarity'].values).T

# Compute the average weighted rating for each book
movie_rec = weighted_ratings.mean(axis=0).reset_index()
movie_rec.columns = ['movie_id', 'weighted_avg_rating']

# Sort recommendations by highest predicted rating
movie_rec = movie_rec.sort_values(by='weighted_avg_rating', ascending=False)

movie_rec
```
