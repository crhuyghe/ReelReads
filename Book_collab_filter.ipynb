```python
import pandas as pd
import random
from sklearn.metrics.pairwise import cosine_similarity


# Load the books dataset
books_file_path = "data/cleaned_books_5000.csv"
books_df = pd.read_csv(books_file_path)

```

```python
ratings_df = pd.read_csv("data/ratings_trimmed_book.csv")
ratings_df = ratings_df.bookId.unique()
ratings_df.tolist()
type(ratings_df)
```

```python
# i want to create a new books_df that takes the books_df and removes the ID that are not in the ratings_df list
books_df = books_df[books_df['Id'].isin(ratings_df)]
books_df.Id.nunique()
```

```python
magic_keywords = ["magic", "wizard", "witch", "sorcery", "spell", "fantasy", "supernatural", "mage"]
magic_books_df = books_df[books_df["Description"].str.contains('|'.join(magic_keywords), case=False, na=False)]

# Select a few magic-related books
selected_magic_books = magic_books_df.sample(n=25, random_state=42)  # Select 5 random books
selected_book_ids = selected_magic_books["Id"].astype(str).tolist()
selected_book_ids
```

```python
# Filter books that contain magical elements
magic_keywords = ["magic", "wizard", "witch", "sorcery", "spell", "fantasy", "supernatural", "mage"]
magic_books_df = books_df[books_df["Name"].str.contains('|'.join(magic_keywords), case=False, na=False)]

# Select a few magic-related books
selected_magic_books = magic_books_df.sample(n=25, random_state=42)  # Select 5 random books
selected_book_ids = selected_magic_books["Id"].astype(str).tolist()
selected_book_ids
```

```python
new_user_vector = pd.DataFrame(0, index=[0], columns=user_item_matrix.columns, dtype=float)
for book_id in selected_book_ids:
    new_user_vector.loc[0, book_id] = round(random.uniform(4.0, 5.0), 2)
new_user_vector
```

```python
# find columns in row 1 where the value isn't 0
non_zero_indices = new_user_vector.iloc[0][new_user_vector.iloc[0] != 0].index
new_user_vector[non_zero_indices]
```

```python
# Load the user-item matrix
user_item_matrix_path = "data/als_book_ratings.csv"
user_item_matrix = pd.read_csv(user_item_matrix_path)
user_item_matrix
```

```python
# add the new_user_vector as a new row at the begining in teh user_item_matrix and call it updated_user_item_matrix
updated_user_item_matrix = pd.concat([new_user_vector, user_item_matrix], ignore_index=True)
updated_user_item_matrix
```

```python
new_user_vector_values = new_user_vector.values
cosine_sim = cosine_similarity(new_user_vector_values, updated_user_item_matrix.values)
cosine_sim
```

```python
# Convert similarity scores to DataFrame
similarity_df = pd.DataFrame({
    'userId': updated_user_item_matrix.index,
    'cosine_similarity': cosine_sim.flatten()
}).sort_values(by='cosine_similarity', ascending=False)

similarity_df
```

```python
similarity_df = similarity_df[similarity_df['userId'] != test_user_id]
similarity_df
```

```python
top_similar_users = similarity_df.head(10)['userId'].tolist()
top_similar_users
```

```python
# Find books rated by similar users but not by the test user
test_user_rated_books = set(updated_user_item_matrix.loc[test_user_id][updated_user_item_matrix.loc[test_user_id] > 0].index)
similar_users_ratings = updated_user_item_matrix.loc[top_similar_users]
similar_users_ratings
```

```python
# Find books that similar users rated but the test user hasn't rated
recommendation_candidates = similar_users_ratings.drop(columns=test_user_rated_books, errors='ignore')

# Calculate average ratings of these books from similar users
recommendations = recommendation_candidates.mean().reset_index()
recommendations.columns = ['book_id', 'avg_rating']

# Sort recommendations by highest average rating
recommendations = recommendations.sort_values(by='avg_rating', ascending=False)
recommendations
```
